# What is the average number of words in movie titles between the year 2000-2005

```python
df1 = df[(df['release_year'] >= 2000 ) & (df['release_year'] <= 2005)]
```

```python
df1.head()
```

```python
df1['average_number_of_words'] = df1.original_title.apply(lambda x: len(x.split()))
```

```python
df1.head()
```

```python
df1['average_number_of_words'].mean()
```

# What is the most common genre for the Vin Diesel and Emma Watson movies?

```python
df['genres'] = df.genres.str.split('|')
```

```python
df['cast'] = df.cast.str.split('|')
```

```python
df1 = df.explode('cast')
```
